```python
# import library 

import pandas as pd
```

```python
# import dataset

df = pd.read_csv("netflix.csv")
```

```python
# melihat 5 data teratas 
df.head(5)
```

```python
# melihat kolom-kolom pada tabel
df.columns
```

```python
# melihat tipe data pada setiap kolom di tabel
df.info()
```

```python
# melihat summary data
df.describe()
```

```python
# melihat data unik
df["rating"].unique()
```

```python
# melihat frekuensi data
df["type"].value_counts()
```

```python
# melihat data pada spesifik column
df["title"]
```

```python
# mengambil data berdasarkan kondisi
```

```python
# menambah / mengubah Kolom
df['title'] = df['title'] + " - " + df["release_year"].astype(str)
df['title']
```

```python
# hapus kolom
df = df.drop('description', axis=1)
df.columns
```

```python
# sorting 
df.sort_values(by='release_year', ascending=True)
```

```python
# bar chart
import matplotlib.pyplot as plt

country_counts = df["country"].value_counts().head(10)
country_counts.plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Top 10 Countries in Netflix Dataset")
plt.xticks(rotation=45)
plt.show()
```

```python
# line chart

release_trend = df["release_year"].value_counts().sort_index()
release_trend.plot(kind='line')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Trend of Title Releases Over the Years")
plt.xticks(rotation=45)
plt.show()
```

```python
# pie chart

type_counts = df["type"].value_counts()
plt.figure()
type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribution of Title Types")
plt.show()
```
